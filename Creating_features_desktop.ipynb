```python
def get_word_count(processed_data):
    words_in_pipeline = []
    word_list = []
    
    delimiter_keys = ['SPACE', '.', ',', 'RETURN']

    for row_idx in range(len(processed_data)):
        curr_key = processed_data[row_idx][1]

        if(curr_key in delimiter_keys):
            if(len(words_in_pipeline)>0):
                key_word = ''
                for char in words_in_pipeline:
                    key_word=key_word+str(char)
                    
                word_list.append(key_word)
                
            words_in_pipeline = []
            continue
            
        if(curr_key=='BACKSPACE'):
            words_in_pipeline = words_in_pipeline[:-1]
            continue
        
        words_in_pipeline.append([curr_key])
        
    return word_list
```

```python
data_dir_desktop = 'C:/Research Activities/Datasets/BB-MAS_Dataset/Desktop_data/Desktop_nonconventional_100_latest/'

User = []
sample = []
RSA_count_list = []
LSA_count_list = []

user_files_desktop = os.listdir(data_dir_desktop)
for i in tqdm(range(len(user_files_desktop))):
        user_file_desktop = user_files_desktop[i]
        curr_user_ind_desktop = int(user_file_desktop[user_file_desktop.find('_')+1:user_file_desktop.find('@')])
        curr_user_sample = int(user_file_desktop[user_file_desktop.rfind('_')+1:user_file_desktop.find('.')])
        data_frame_desktop = pd.read_csv(data_dir_desktop+user_file_desktop, header=0)
        data_frame_desktop.iloc[:,2] =  pd.to_datetime(data_frame_desktop.iloc[:,2])
        data_frame_desktop.iloc[:,3] =  pd.to_datetime(data_frame_desktop.iloc[:,3])
        
     
        
        Total_shifts=0
        RSA = 0
        LSA = 0
        for row_id in range(len(data_frame_desktop)-1):
            if (data_frame_desktop.iloc[row_id,1]=='RSHIFT'):
                Total_shifts += 1
                RSA += 1
                       
            if data_frame_desktop.iloc[row_id,1]=='LSHIFT':
                Total_shifts += 1
                LSA += 1
                
            
        RSA_ratio = (RSA/(1 if Total_shifts==0 else Total_shifts))*100
        LSA_ratio = (LSA/(1 if Total_shifts==0 else Total_shifts))*100
        
        User.append(curr_user_ind_desktop) 
        sample.append(curr_user_sample)
        RSA_count_list.append(RSA_ratio)
        LSA_count_list.append(LSA_ratio)
    
        
final_df_shift_ratio = pd.DataFrame(list(zip(User,sample,RSA_count_list,LSA_count_list)),
               columns =['User','sample','RSA_ratio','LSA_ratio']) 
```
